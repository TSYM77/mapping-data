```python
import csv
import json
import pandas as pd

input_file = "sample_data/wine-ratings.csv"
dataset = pd.read_csv(input_file)
dataframe1 = pd.DataFrame(dataset)
cols = [1,4,5,3]
select_fields = dataframe1[dataframe1.columns[cols]]
#select_fields.head()
temp_df = select_fields.loc[(select_fields['variety'] == 'Red Wine') & 
                         (select_fields['region'] == 'California') & 
                         (select_fields['rating'] >= 90)]
temp_df.head(n=200)

result = temp_df.to_json(r'sample_data/final_wine_list.json', orient="index")

```
